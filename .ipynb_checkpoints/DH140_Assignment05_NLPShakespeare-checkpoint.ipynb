## Macbeth

Make a list of the words in Macbeth by tokenizing `macbeth_string`:

```python
import nltk
from nltk.tokenize import word_tokenize

# Tokenize the macbeth_string to create a list of words
macbeth_words = word_tokenize(macbeth_string)

# Print the first 10 words to verify
print(macbeth_words[:10])
```

Next, make a list of positive words and a list of negative words using the `polarity_scores` method of VADER's SentimentIntensityAnalyzer.  To do this:
* Write a for loop that iterates over your Macbeth word list
  * If the compound score of a word is positive, add the word to the positive-word list
  * If the compound score of a word is negative, add the word to the negative-word list
  * If the compound score is 0, the word should not be added to either list

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize VADER's SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

# Create empty lists for positive and negative words
positive_words = []
negative_words = []

# Iterate over the tokenized words from "Macbeth"
for word in macbeth_words:
    score = sid.polarity_scores(word)['compound']
    if score > 0:
        positive_words.append(word)
    elif score < 0:
        negative_words.append(word)

# Print the first 10 words in each list to verify
print("Positive words:", positive_words[:10])
print("Negative words:", negative_words[:10])
```

1. Find the 15 most common negative words in Macbeth and the 15 most common positive words
    * Note that if you have a list variable called `mylist`, you can get the frequencies of words in `mylist` by using `nltk.FreqDist(mylist)`, and you can get the `n` most common words with `nltk.FreqDist(mylist).most_common(n)`.
2. Make a horizontal bar plot similar to the plots in the following image

(Plot taken from a [2018 post by Peer Chirstensen](https://web.archive.org/web/20230129014418/https://peerchristensen.netlify.app/post/fair-is-foul-and-foul-is-fair-a-tidytext-entiment-analysis-of-shakespeare-s-tragedies/))



```python
import matplotlib.pyplot as plt
import nltk

# Calculate the frequencies of words in the positive and negative word lists
positive_freq = nltk.FreqDist(positive_words)
negative_freq = nltk.FreqDist(negative_words)

# Find the 15 most common words in each list
top_15_positive = positive_freq.most_common(15)
top_15_negative = negative_freq.most_common(15)

# Separate the words and their frequencies for plotting
positive_words, positive_counts = zip(*top_15_positive)
negative_words, negative_counts = zip(*top_15_negative)

# Plot the 15 most common positive words
plt.figure(figsize=(10, 6))
plt.barh(positive_words, positive_counts, color='green')
plt.xlabel('Frequency')
plt.title('15 Most Common Positive Words in Macbeth')
plt.gca().invert_yaxis()
plt.show()

# Plot the 15 most common negative words
plt.figure(figsize=(10, 6))
plt.barh(negative_words, negative_counts, color='red')
plt.xlabel('Frequency')
plt.title('15 Most Common Negative Words in Macbeth')
plt.gca().invert_yaxis()
plt.show()
```

There are a lot of similarities between the Macbeth negative/positive words and the words in the other bar plots. 

The word list we created from Macbeth did not have the stopwords removed.
* Repeat the above analysis, but do so for a list of Macbeth words in which the stopwords have been removed.
* What changes?

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Get the list of stopwords
stop_words = set(stopwords.words('english'))

# Tokenize Macbeth text
macbeth_words = word_tokenize(macbeth_string)

# Remove stopwords
macbeth_words_no_stopwords = [word for word in macbeth_words if word.isalpha() and word not in stop_words]

# Initialize VADER SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Lists for positive and negative words
positive_words = []
negative_words = []

# Analyze the sentiment of each word
for word in macbeth_words_no_stopwords:
    score = sia.polarity_scores(word)['compound']
    if score > 0:
        positive_words.append(word)
    elif score < 0:
        negative_words.append(word)

# Find the 15 most common positive and negative words
positive_word_freq = nltk.FreqDist(positive_words).most_common(15)
negative_word_freq = nltk.FreqDist(negative_words).most_common(15)

# Plot the results
def plot_word_freq(word_freq, title):
    words, frequencies = zip(*word_freq)
    plt.figure(figsize=(10, 8))
    plt.barh(words, frequencies, color='skyblue')
    plt.xlabel('Frequency')
    plt.title(title)
    plt.gca().invert_yaxis()
    plt.show()

plot_word_freq(positive_word_freq, '15 Most Common Positive Words in Macbeth (Stopwords Removed)')
plot_word_freq(negative_word_freq, '15 Most Common Negative Words in Macbeth (Stopwords Removed)')
```

The words that appear in the lists are more meaningful and content-specific, as the stopwords (which are generally neutral and not content-rich) are excluded.

* Repeat the above analysis one more time -> in addition to removing the stopwords, use the Porter stemmer to get a list of stemmed words before you calculate the word frequencies.
* What words in your plot have now been counted together?

```python
from nltk.stem import PorterStemmer

# Initialize Porter Stemmer
porter_stemmer = PorterStemmer()

# Remove stopwords and stem words
macbeth_words_no_stopwords_stemmed = [porter_stemmer.stem(word) for word in macbeth_words if word.isalpha() and word not in stop_words]

# Initialize VADER SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Lists for positive and negative words
positive_words_stemmed = []
negative_words_stemmed = []

# Analyze the sentiment of each word
for word in macbeth_words_no_stopwords_stemmed:
    score = sia.polarity_scores(word)['compound']
    if score > 0:
        positive_words_stemmed.append(word)
    elif score < 0:
        negative_words_stemmed.append(word)

# Find the 15 most common positive and negative words
positive_word_freq_stemmed = nltk.FreqDist(positive_words_stemmed).most_common(15)
negative_word_freq_stemmed = nltk.FreqDist(negative_words_stemmed).most_common(15)

# Plot the results
def plot_word_freq(word_freq, title):
    words, frequencies = zip(*word_freq)
    plt.figure(figsize=(10, 8))
    plt.barh(words, frequencies, color='skyblue')
    plt.xlabel('Frequency')
    plt.title(title)
    plt.gca().invert_yaxis()
    plt.show()

plot_word_freq(positive_word_freq_stemmed, '15 Most Common Positive Words in Macbeth (Stopwords Removed & Stemmed)')
plot_word_freq(negative_word_freq_stemmed, '15 Most Common Negative Words in Macbeth (Stopwords Removed & Stemmed)')
```

By using the Porter stemmer, words that share the same root form are counted together. For example:  
"murder" and "murdered" are both stemmed to "murder."


## Romeo and Juliet

We are now going to plot the evolution of sentiment in Romeo and Juliet.

We are going to make a couple different plots.

First:
* Write a for loop that iterates over the scenes in Romeo and Juliet (`r_j_scenes`), and for each scene:
  * Make a list of the scene's words by tokenizing the string
  * Sum the compound sentiment scores of every word in the scene (using VADER's SentimentIntensityAnalyser's polarity_scores method to get the compound sentiment score for each word).
* Make a bar plot that has the scene number on the horizontal axis and the sentiment scores as the bar heights.
* Comment on its resemblance to the Romeo and Juliet plot above.

```python
# Initialize VADER SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# List to store sentiment scores for each scene
scene_sentiments = []

# Iterate over each scene
for scene in r_j_scenes:
    # Tokenize the scene into words
    scene_words = word_tokenize(scene)
    
    # Sum the compound sentiment scores of each word in the scene
    scene_sentiment = sum(sia.polarity_scores(word)['compound'] for word in scene_words)
    
    # Append the sentiment score of the scene to the list
    scene_sentiments.append(scene_sentiment)

# Create a bar plot with scene numbers on the horizontal axis and sentiment scores as bar heights
plt.figure(figsize=(14, 8))
plt.bar(range(1, len(scene_sentiments) + 1), scene_sentiments, color='skyblue')
plt.xlabel('Scene Number')
plt.ylabel('Summed Sentiment Score')
plt.title('Sentiment Evolution in Romeo and Juliet')
plt.xticks(range(1, len(scene_sentiments) + 1))  # Ensure every scene number is labeled
plt.show()
```

The two graphs show similar shapes and form when compared to one another. There are peaks and troughs at key scenes corresponding to significant events (e.g., romantic encounters, conflicts, or deaths), though the exact positions and heights of the bars.


Second:
* Write a for loop that iterates over the scenes in Romeo and Juliet (`r_j_scenes`), and for each scene:
  * Make a list of the scene's words by tokenizing the string
  * Calculate the total scene sentiment score by **adding 1 to the score for each positive word (compound_score > 0) and substracting 1 from the score for each negative word (compound_score < 0)**
    * This total is the difference between the positive and negative word counts
* Make a bar plot that has the scene number on the horizontal axis and the sentiment scores as the bar heights.
* Does this change the plot relative to the previous method of calculating the total scene sentiment score?

```python
# Initialize VADER SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# List to store sentiment scores for each scene
scene_sentiments = []

# Iterate over each scene
for scene in r_j_scenes:
    # Tokenize the scene into words
    scene_words = word_tokenize(scene)
    
    # Initialize sentiment score for the scene
    scene_sentiment = 0
    
    # Calculate the sentiment score for each word in the scene
    for word in scene_words:
        compound_score = sia.polarity_scores(word)['compound']
        if compound_score > 0:
            scene_sentiment += 1
        elif compound_score < 0:
            scene_sentiment -= 1
    
    # Append the sentiment score of the scene to the list
    scene_sentiments.append(scene_sentiment)

# Create a bar plot with scene numbers on the horizontal axis and sentiment scores as bar heights
plt.figure(figsize=(14, 8))
plt.bar(range(1, len(scene_sentiments) + 1), scene_sentiments, color='skyblue')
plt.xlabel('Scene Number')
plt.ylabel('Sentiment Score (Positive - Negative)')
plt.title('Sentiment Evolution in Romeo and Juliet')
plt.xticks(range(1, len(scene_sentiments) + 1))  # Ensure every scene number is labeled
plt.show()
```

The second method uses a more straightforward approach by simply counting positive and negative words and calculating the difference, whereas the first method sums the compound sentiment scores. Both methods show similar overall trends in sentiment changes across the scenes, but the magnitude and specific values of the sentiment scores differ.

Third plot:

* Write a for loop that iterates over the scenes in Romeo and Juliet (`r_j_scenes`), and for each scene:
  * Make a list of the scene's words by tokenizing the string
  * Calculate the total scene sentiment score by **adding 1 for each positive word and substracting 1 for each negative word, where the positive/negative word is assessed by its presence in the `positive_words` and `negative_words` sets.**
    * This total is the difference between the positive and negative word counts
* Make a bar plot that has the scene number on the horizontal axis and the sentiment scores as the bar heights.
* Does this change the plot relative to the previous plots?

```python
from nltk.corpus import opinion_lexicon

# Make a set of positive words and a set of negative words
positive_words = set(opinion_lexicon.positive())
negative_words = set(opinion_lexicon.negative())

# Example of checking whether a word is positive or negative
testword = 'love'
if testword in positive_words:
    print(testword + ' is positive')
elif testword in negative_words:
    print(testword + ' is negative')
else:
    print(testword + ' is neither positive nor negative')

# List to store sentiment scores for each scene
scene_sentiments = []

# Iterate over each scene
for scene in r_j_scenes:
    # Tokenize the scene into words
    scene_words = word_tokenize(scene)
    
    # Initialize sentiment score for the scene
    scene_sentiment = 0
    
    # Calculate the sentiment score for each word in the scene
    for word in scene_words:
        if word in positive_words:
            scene_sentiment += 1
        elif word in negative_words:
            scene_sentiment -= 1
    
    # Append the sentiment score of the scene to the list
    scene_sentiments.append(scene_sentiment)

# Create a bar plot with scene numbers on the horizontal axis and sentiment scores as bar heights
plt.figure(figsize=(14, 8))
plt.bar(range(1, len(scene_sentiments) + 1), scene_sentiments, color='skyblue')
plt.xlabel('Scene Number')
plt.ylabel('Sentiment Score (Positive - Negative)')
plt.title('Sentiment Evolution in Romeo and Juliet (Opinion Lexicon)')
plt.xticks(range(1, len(scene_sentiments) + 1))  # Ensure every scene number is labeled
plt.show()
```

Using the opinion_lexicon changes how positive and negative words are identified. This lexicon has different sets of positive and negative words compared to the VADER lexicon. The overall trend of sentiment changes across the scenes and  the specific values and the appearance of the sentiment scores differ due to the differences in the lexicons used.


Fourth plot:

* Take the entire string of Romeo and Juliet (`r_j_string`) and tokenize it into words.
* Calculate a total sentiment score for each consecutive group of 500 words.
  * Do this as immediately above by adding 1 for each positive word and substracting 1 for each negative word, where the positive/negative word is assessed by its presence in the `positive_words` and `negative_words` sets.
* Make a bar plot that has the group number (0,1,2,3,...) on the horizontal axis and the sentiment scores as the bar heights.
* Does this change the plot relative to the previous plots?

```python
# Tokenize the entire Romeo and Juliet string into words
r_j_words = word_tokenize(r_j_string)

# List to store sentiment scores for each group of 500 words
group_sentiments = []

# Iterate over the words in chunks of 500
chunk_size = 500
for i in range(0, len(r_j_words), chunk_size):
    # Get the current chunk of 500 words
    word_chunk = r_j_words[i:i + chunk_size]
    
    # Initialize sentiment score for the chunk
    chunk_sentiment = 0
    
    # Calculate the sentiment score for each word in the chunk
    for word in word_chunk:
        if word in positive_words:
            chunk_sentiment += 1
        elif word in negative_words:
            chunk_sentiment -= 1
    
    # Append the sentiment score of the chunk to the list
    group_sentiments.append(chunk_sentiment)

# Create a bar plot with group numbers on the horizontal axis and sentiment scores as bar heights
plt.figure(figsize=(14, 8))
plt.bar(range(len(group_sentiments)), group_sentiments, color='skyblue')
plt.xlabel('Group Number (0, 1, 2, 3, ...)')
plt.ylabel('Sentiment Score (Positive - Negative)')
plt.title('Sentiment Evolution in Romeo and Juliet (Opinion Lexicon, 500-word Chunks)')
plt.show()
```

This method provides a finer granularity of sentiment analysis, capturing sentiment changes more frequently across the entire text of "Romeo and Juliet". While the overall sentiment trend might still be similar to the scene-based plots, the increased resolution allows for the detection of more subtle sentiment shifts within and between scenes.
